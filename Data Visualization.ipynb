```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a dataset
np.random.seed(0)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = {
    'Category A': np.random.normal(loc=10, scale=2, size=100),
    'Category B': np.random.normal(loc=15, scale=3, size=100),
    'Category C': np.random.normal(loc=20, scale=4, size=100),
    'Category D': np.random.normal(loc=25, scale=5, size=100),
}

# Convert to DataFrame
data = pd.DataFrame({cat: val for cat, val in values.items()}).melt(var_name='Category', value_name='Value')

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=data, palette='Set2')

# Adding titles and labels
plt.title('Distribution of Numerical Values Across Categories')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show grid
plt.grid(axis='y')

# Show plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(0)

# Generate synthetic data
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = x1 * 0.8 + np.random.normal(0, 0.1, n_samples)  # Correlated with x1
x3 = np.random.rand(n_samples)
x4 = x3 * 0.5 + np.random.normal(0, 0.2, n_samples)  # Correlated with x3

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar=True, linewidths=.5)

# Adding titles
plt.title('Correlation Matrix Heatmap')

# Show plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)  # 100 values from -5 to 5
y = np.linspace(-5, 5, 100)  # 100 values from -5 to 5
x, y = np.meshgrid(x, y)  # Create a grid of x and y values

# Calculate z values based on the function z = sin(sqrt(x² + y²))
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(df['X'].values.reshape(x.shape), df['Y'].values.reshape(y.shape), df['Z'].values.reshape(z.shape), cmap='viridis')

# Adding titles and labels
ax.set_title('3D Surface Plot of z = sin(sqrt(x² + y²))')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(25)

# Generate the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'],
            s=df['Population'] * 10,  # Bubble size proportional to population
            alpha=0.5,
            c='blue',
            edgecolors='w',
            linewidth=2)

# Adding titles and labels
plt.title('Bubble Chart: Population vs GDP by Country')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')

# Annotate each bubble with the country name
for i in range(len(df)):
    plt.annotate(df['Country'][i],
                 (df['GDP'][i], df['Population'][i]),
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha='center')

# Show grid
plt.grid(True)

# Show plot
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output to display in a Jupyter notebook
output_notebook()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Calculate the sine of x

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add the sine wave to the figure
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

# Prepare the output to display in a Jupyter notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 10  # Random x values from 0 to 10
y = np.random.rand(n) * 10  # Random y values from 0 to 10
sizes = np.random.randint(10, 100, n)  # Random sizes for the markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Random colors

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a Bokeh figure
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Add scatter points to the figure
p.scatter('x', 'y', size='sizes', color='colors', source=data, legend_field='colors', fill_alpha=0.6)

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output to display in a Jupyter notebook
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts',
           height=400, width=600)

# Add bars to the figure
p.vbar(x=fruits, top=counts, width=0.5, color='blue')

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Normally Distributed Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           height=400, width=600)

# Add bars to the figure
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='white')

# Show the plot
show(p)
```
